```python
import pandas as pd
movie_data = pd.read_csv('/content/top10K-TMDB-movies.csv')
movie_data.head()
```

```python
movie_data.info() #type of data
```

```python
movie_data.isnull().sum()
```

```python
movie_data.columns
```

```python
movie_data = movie_data[['id', 'title', 'overview', 'genre']]
movie_data
```

```python
movie_data['tags'] = movie_data['overview']+' '+movie_data['genre']
movie_data
```

```python
new_data = movie_data.drop(columns=['overview', 'genre'])
new_data
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv = CountVectorizer(max_features=10000, stop_words='english')
```

```python
vector=cv.fit_transform(new_data['tags'].values.astype('U')).toarray()
vector.shape
```

```python
from sklearn.metrics.pairwise import cosine_similarity
similarity = cosine_similarity(vector)
```

```python
new_data[new_data['title']=='The Godfather'].index[0]
#this is used to get the index of the movie
```

```python
similarity_score = sorted(list(enumerate(similarity[2])),reverse=True,key=lambda vector: vector[1])
for i in similarity_score[0:5]:
  print(new_data.iloc[i[0]].title)
  #the 0 index stands for the ID column in the new_data dataframe
```

```python
def recommend_movie(movies):
  index = new_data[new_data['title']==movies].index[0]
  similarity_score = sorted(list(enumerate(similarity[index])),reverse=True,key=lambda vector: vector[1])
  for i in similarity_score[0:5]:
    print(new_data.iloc[i[0]].title)
```

```python
recommend_movie('Batman')
```
